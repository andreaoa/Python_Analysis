```python
#Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
plt.style.use('ggplot')


#Remove limitation on number of columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('Online_Retail.csv', sep=';', decimal=',')
df.head()
```

```python
#Vis√£o geral dos dados
df.info()
```

```python
#Ver se tem valores duplicados
df.duplicated().sum()
```

```python
#Removendo os duplicados
df1 = df.loc[~df.duplicated()].reset_index(drop=True).copy()
df1.head()
```

```python
df1['InvoiceDate'] = pd.to_datetime(df1['InvoiceDate'], format='%Y/%m/%d')
df1['CustomerID'] = df['CustomerID'].astype('str')
df1['TotalSale'] = (df['Quantity'] * df['UnitPrice'])
df1.dtypes
```

```python
df1.groupby(['Country']).agg(np.sum).sort_values('Quantity', ascending=False)
```

```python
#Produtos Mais vendidos em quantidade
df1.groupby(['Description']).agg(np.sum).sort_values('Quantity', ascending=False).head()
```

```python
#Produtos mais vendidos em receita
df1.groupby(['Description']).agg(np.sum).sort_values('TotalSale', ascending=False).head()
```
